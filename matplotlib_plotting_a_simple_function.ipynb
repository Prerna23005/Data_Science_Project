```python
df=pd.read_csv('/content/batsman_season_record.csv')
df
```

```python
plt.plot(df['2015'],df['batsman'])
plt.plot(df['2016'],df['batsman'])
plt.plot(df['2017'],df['batsman'])
plt.show()
```

```python
years = ['2015', '2016', '2017']

# Get V Kohli's runs for these years (V Kohli is at index 4)
v_kohli_runs = df.loc[4, years]

# Get RG Sharma's runs for these years (RG Sharma is at index 3)
rg_sharma_runs = df.loc[3, years]

plt.plot(years, v_kohli_runs, label='V Kohli')
plt.plot(years, rg_sharma_runs, label='RG Sharma')
plt.title('V Kohli vs RG Sharma Runs Comparison (2015-2017)')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.legend()
plt.show()
```

```python
years = ['2015', '2016', '2017']

v_kohli_runs = df.loc[4, years] # V Kohli is at index 4
rg_sharma_runs = df.loc[3, years] # RG Sharma is at index 3

plt.plot(years, v_kohli_runs, color='#D9F10F', label='V Kohli')
plt.plot(years, rg_sharma_runs, color='#FC00D6', label='RG Sharma')
plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend()
plt.show()
```

```python
years = ['2015', '2016', '2017']

v_kohli_runs = df.loc[4, years] # V Kohli is at index 4
rg_sharma_runs = df.loc[3, years] # RG Sharma is at index 3

plt.plot(years, v_kohli_runs, color='#D9F10F', label='V Kohli',linewidth=3,linestyle='solid',marker='D',markersize=15,)
plt.plot(years, rg_sharma_runs, color='#FC00D6', label='RG Sharma',linewidth=4,linestyle='dashdot',marker='o',markersize=10,)
plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend()
plt.grid(True)
plt.show()
```
